#### 1.
The equality constraint was defined as:
$$
y_t(\theta^Tx_t+\mu)<0
$$
so we would like to minimise this term to find the optimal solution.
$$
\min_{(\theta,\mu),(z_i\in \Reals)} \sum^n_{t=1}\ell(z_t)+\frac{\alpha}{2}||\theta||^2_2\\
s.t.\\
z_t= y_t(\theta^Tx_t+ \mu)
$$
To obtain the lagarangian function we introduce lagrange multipliers to the equality constraints. 
Bringing in the equality constraint and multiplying by the lagrange multiplier $\lambda_t$ associated with the ith equality constraint.
$$
\mathcal{L}(\theta,\mu,\{z_i\in \Reals\},\lambda) = \sum^n_{t=1}\ell(z_t)+\frac{\alpha}{2}||\theta||^2_2 + \sum^n_{t=1}\lambda_t(z_t - y_t(\theta^Tx_t+ \mu))
$$

Slater's condition is a condition for strong duality. It states that if the problem is convex and there exists some feasible point in the interior of the domain, then strong duality holds.

The problem in (4) is convex because the objective function is the sum of a convex function $\ell(z)$ and a quadratic function, both of which are convex. The constraints are linear and therefore both convex or concave.

For any $\theta$ and $\mu$, we can always find $z_i = y_i(\theta^T x_i + \mu)$ that satisfies the constraints wich means there exists some feasible point in the interior of the domain. Therefore, Slater's condition is satisfied for this problem and strong duality holds. This means we can solve the primal problem by solving its dual.

#### 2.
The Lagrangian min-max optimization problem can be written as:

$$
\min_{(\theta, \mu), \{z_i \in R\}} \max_{\lambda \geq 0} L((\theta, \mu), \{z_i \in R\},\lambda)
$$

where $L((\theta, \mu), \{z_i \in R\},\lambda)$ is the Lagrangian function given by:

$$
L((\theta, \mu), \{z_i \in R\},\lambda) = \sum_{i=1}^{n} \ell(z_i) + \frac{\alpha}{2} \|\theta\|^2 + \sum_{i=1}^{n} \lambda_i (y_i(\theta^T x_i + \mu) - z_i)
$$

By interchanging min and max, we get the dual problem:

$$
\max_{\lambda \geq 0} \min_{(\theta, \mu), \{z_i \in R\}} L((\theta, \mu), \{z_i \in R\},\lambda)
$$

For a convex $\ell$, we can solve the inner minimization problem by setting its derivatives with respect to $\theta$, $\mu$, and $z_i$ to zero. The derivatives are as follows:

Taking the derivative with respect to $\theta$ gives:

$$
\frac{\partial L}{\partial \theta} = \alpha \theta - \sum_{i=1}^{n} \lambda_i y_i x_i = 0
$$

Taking the derivative with respect to $\mu$ gives:

$$
\frac{\partial L}{\partial \mu} = - \sum_{i=1}^{n} \lambda_i y_i = 0
$$

Taking the derivative with respect to $z_i$ gives:

$$
\frac{\partial L}{\partial z_i} = \ell'(z_i) - \lambda_i = 0
$$

Rearranging for the target variables $\theta$, $\mu$, and $z_i$ gives us:

$$
\theta = -\frac{1}{\alpha} \sum_{i=1}^{n} \lambda_i y_i x_i
$$

$$
\sum_{i=1}^{n} \lambda_i y_i = 0
$$

and

$$
z_i = \ell'^{-1}(-\lambda_i)
$$

where $\ell'^{-1}$ is the inverse of the derivative of $\ell$. 

Substituting these into the Lagrangian gives the dual problem:

$$
L((\theta, \mu), \{z_i \in R\},\lambda) = \sum_{i=1}^{n} \ell(\ell'^{-1}(-\lambda_i)) + \frac{\alpha}{2} \left\| \frac{1}{\alpha} \sum_{i=1}^{n} \lambda_i y_i x_i \right\|^2 + \sum_{i=1}^{n} \lambda_i (y_i(\theta^T x_i + \mu) - \ell'^{-1}(-\lambda_i))
$$

The term $\sum_{i=1}^{n} \lambda_i (y_i(\theta^T x_i + \mu) - \ell'^{-1}(-\lambda_i))$ simplifies to zero because of the constraints $z_i = \ell'^{-1}(-\lambda_i)$ and $\sum_{i=1}^{n} \lambda_i y_i = 0$.

The term $\ell(\ell'^{-1}(-\lambda_i))$ simplifies to $\ell^*(\lambda_i)$ by the definition of the convex conjugate. 

The term $\frac{\alpha}{2} \left\| \frac{1}{\alpha} \sum_{i=1}^{n} \lambda_i y_i x_i \right\|^2$ simplifies to $\frac{1}{2\alpha} \left\| \sum_{i=1}^{n} \lambda_i y_i x_i \right\|^2$ by removing the common factor of $\frac{1}{\alpha}$.

$$
\min_{\{\lambda_i \in R\}} \sum_{i=1}^{n} \ell^*(\lambda_i) + \frac{1}{2\alpha} \left\| \sum_{i=1}^{n} \lambda_i y_i x_i \right\|^2
\\
\text{s.t.} 
\\
\sum_{i=1}^{n} \lambda_i y_i = 0
$$

where $\ell^*$ is the convex conjugate of $\ell$. This is the same as the problem in (5). The convex conjugate $\ell^*$ of a function $\ell$ is defined as:

$$
\ell^*(\lambda) = \sup_{z \in dom(\ell)} \{\lambda z - \ell(z)\}
$$

##### a)
The logistic loss function is given by $\ell(z) = \log(1 + e^z)$. The convex conjugate of the logistic loss, denoted as $\ell^*(\lambda)$, can be calculated as follows:

$$
\ell^*(\lambda) = \sup_{z \in dom(\ell)} \{\lambda z - \ell(z)\}
$$

To find the supremum, we take the derivative of the inner expression with respect to $z$ and set it equal to zero:

$$
\frac{d}{dz} (\lambda z - \log(1 + e^z)) = \lambda - \frac{e^z}{1 + e^z} = 0
$$

Solving for z:
$$\lambda = \frac{e^z}{1 + e^z} $$
$$\lambda(1 + e^z) = e^z $$
$$\lambda = e^z (1 - \lambda) $$
$$ e^z = \frac{\lambda}{1 - \lambda} $$
$$z = \log(\lambda) - \log(1-\lambda)$$

Substituting this back into the expression gives:

$$
\ell^*(\lambda) = \lambda \log(\lambda) + (1-\lambda) \log(1-\lambda)
$$

This is defined for $\lambda \in (0,1)$.

##### b)
The hinge loss function is given by $\ell(z) = \max(0, z)$. The convex conjugate of the hinge loss, denoted as $\ell^*(\lambda)$, can be calculated as follows:

$$
\ell^*(\lambda) = \sup_{z \in dom(\ell)} \{\lambda z - \ell(z)\}
$$

The supremum is achieved at $z = 0$ for $\lambda < 0$, and at $z = \lambda$ for $\lambda \geq 0$. 

Substituting these back into the expression gives:

$$
\ell^*(\lambda) = 
\begin{cases} 
0 & \text{if } \lambda < 0 \\
\frac{\lambda^2}{2} & \text{if } \lambda \geq 0 
\end{cases}
$$

This is defined for all $\lambda \in R$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def logistic_conjugate(x):
    return x * np.log(x) + (1 - x) * np.log(1 - x)

def hinge_conjugate(x):
    return np.where(x < 0, 0, x**2 / 2)

# Generate x values
x_logistic = np.linspace(0.01, 0.99, 100)
x_hinge = np.linspace(-2, 2, 100)

# Generate y values
y_logistic = logistic_conjugate(x_logistic)
y_hinge = hinge_conjugate(x_hinge)

# Create the plots
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x_logistic, y_logistic)
plt.title('Convex Conjugate of Logistic Loss')
plt.xlabel('$\lambda$')
plt.ylabel('Convex Conjugate')

plt.subplot(1, 2, 2)
plt.plot(x_hinge, y_hinge)
plt.title('Convex Conjugate of Hinge Loss')
plt.xlabel('$\lambda$')
plt.ylabel('Convex Conjugate')

plt.show()
```

```python
import numpy as np
from sklearn import svm
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm

def get_Y(beta, theta0, X):
    Y = []
    for x_i in X:
        t_i = np.matmul(theta0.T, x_i)
        odds = np.exp(beta*t_i)
        Y.append(np.random.choice(a=[-1, 1], size=1, p= [odds/(1+odds), 1/(1+odds)])[0])
    return Y

def error_calc(alpha,n,d,beta):
    theta0=np.ones(d)
    # Write a code that generates a synthetic dataset X,Y
    # X has n rows (number of data points) and d columns (number of features)
    # and Y is a 1-d array with n elements each being -1 or 1
    # each element of X is standard Gaussian
    # for the i element of Y is randomly generated by the
    # description in the pdf file and parameter alpha
    X = np.random.normal(0, 1, size=(n, d))    
    Y = get_Y(beta, theta0, X)
    Y = np.array(Y)
    # run svm.SVC with parameters (C=1/alpha,kernel='linear')
    # also run Logisticregression with parameter (C=1/alpha)
    # See: https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html
    # See: https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html
    if len(np.unique(Y)) == 2:    
        model_LR = LogisticRegression(C=1/alpha)
        model_LR.fit(X=X, y=Y)
        model_SVC = svm.SVC(C=1/alpha, kernel='linear')
        model_SVC.fit(X=X, y=Y)
    else:
        print('Y has only one class this iteration is ignored')
        return None, None

    # read out the coefficients of the trained models
    coef_LR = model_LR.coef_
    coef_SVM = model_SVC.coef_

    # calculate SQUARE L2 norm of the error between the coefficients of each model
    # and theta0, call the error_LR and error_SVM
    error_LR = np.linalg.norm(coef_LR - theta0) ** 2
    error_SVM = np.linalg.norm(coef_SVM - theta0) ** 2

    return error_LR, error_SVM

def avg_risk_calc(reg,n,d,alpha):
    # write a for loop that calculates the error for 1000 times
    # Although the parametrs reg, n,d,alpha are the same, each error is random
    # because the dataset is generated again
    # Calculate the average errors of 1000 runs 
    # You get MSE_LR and MSE_SVM
    errors_LR = []
    errors_SVC = []
    for _ in range(1000):
        error_LR, error_SVC = error_calc(alpha, n, d, reg)
        if error_LR is None or error_SVC is None:
            continue
        errors_LR.append(error_LR)
        errors_SVC.append(error_SVC)
    MSE_LR = np.mean(errors_LR)
    MSE_SVC = np.mean(errors_SVC)
    return MSE_LR, MSE_SVC

# Now using avg_risk_calc function plot the risk function of alpha
# for different values n,d, reg 
n_values = [100, 500, 1000]
d_values = [2, 10, 20]
alpha_values = [0.1, 2, 5]


fig, axs = plt.subplots(len(n_values), len(d_values), figsize=(20, 20))

for i, n in enumerate(n_values):
    for j, d in enumerate(d_values):
        for alpha in alpha_values:
            MSE_LR_values = []
            MSE_SVM_values = []
            reg_values = np.linspace(0.1, 5, num=10)
            for reg in reg_values:
                print(f'n={n}, d={d} alpha={alpha}, reg={reg},')
                MSE_LR, MSE_SVM = avg_risk_calc(reg, n, d, alpha)
                MSE_LR_values.append(MSE_LR)
                MSE_SVM_values.append(MSE_SVM)
            axs[i, j].plot(reg_values, MSE_LR_values, label=f'Logistic Regression alpha={alpha}')
            axs[i, j].plot(reg_values, MSE_SVM_values, label=f'SVC alpha={alpha}')
            axs[i, j].set_xlabel('beta')
            axs[i, j].set_ylabel('MSE')
            axs[i, j].set_title(f'n={n}, d={d}')
            axs[i, j].legend()
plt.tight_layout()
plt.show()
fig.savefig('Code_2.png')
```

### <center> Q2. (Expectation Maximization Clustering)

#### 1.

For $F(\theta,\theta^{(t)})$ to be a majorizer of $L(\theta)$ at $\theta^{(t)}$ we need 
$$
F(\theta^{(t)},\theta^{(t)}) = L(\theta^{(t)})
$$
and
$$
L(\theta) \leq F(\theta, \theta^{(t)})\text{ for all } \theta
$$  
to be true. 


For the first statment when we apply 
$$
F(\theta^{(t)},\theta^{(t)}) = \sum^n_{i=1} \sum^K_{k=1} \frac{\omega_{i,k}}{2\sigma^2} (\| x_i - \mu_k^{(t)} \|^2_2 - \| x_i - \mu_k^{(t)} \|^2_2) + L(\theta^{(t)})
$$

and sense the the norms cancel eachother when $\theta = \theta^{(t)}$ we are left with
$$
F(\theta^{(t)},\theta^{(t)}) =  L(\theta^{(t)})
$$

Proving the first condition.  






For the second statment we take the equation 

$$
L(\theta) = \sum^n_{i=1} \log (\sum^K_{k=1} e^{ -\frac{\| x_i - \mu_k \|^2_2 }{2\sigma^2}})
$$

and we can apply $\zeta_{i,k}$ and $w_{i,k}$ to get

$$
L(\theta) = \sum^n_{i=1} \log (\sum^K_{k=1}\zeta_{i,k} w_{i,k})
$$

Now we can apply Jensen's inequality and we get

$$
-\log (\sum^K_{k=1}\zeta_{i,k} w_{i,k}) \leq -\sum^K_{k=1}w_{i,k}log(\zeta_{i,k}) 
$$


#### 2.


To find the optimal solution we take the derivative of $F(\theta, \theta^{(t)})$ with respect to $\mu_k$

$$\frac{\partial F(\theta, \theta^{(t)})}{\partial \mu_k} = \sum_{i=1}^{n} \omega_{i,k} (x_i - \mu_k)$$

Setting this equal to zero gives us:

$$\sum_{i=1}^{n} \omega_{i,k} (x_i - \mu_k) = 0$$

and solving for $\mu_k^{t+1}$

$$\mu_k^{t+1} = \frac{\sum^n_{i=1}\omega_{i,k}x_i}{\sum^n_{i=1}\omega_{i,k}}$$



#### 3.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tqdm.notebook as tqdm

def ex_neg_norm(x, mu, sigma):
    return np.exp(-(np.linalg.norm(x - mu)**2) / (2 * sigma**2))

def EM_clustering(X,sigma,K):
    # Number of iterations
    iter_num=10
    # d is the number of features and n is the number of the data points:
    (n,d)=X.shape
    # Initialize the K centroids randomly. Each row is a centroid.
    M=np.random.randn(K,d)
    # A vector that stores the NLL value in different iterations:
    value=np.zeros(iter_num)
    #start the EM loop:
    for t in range(iter_num):
        # Apply the E stape by calculating the omega_{i,k} coefficients
        omega = np.zeros((n, K))
        for i in range(n):
            for j in range(K):
                omega[i, j] = ex_neg_norm(X[i], M[j], sigma)
            omega[i] /= np.sum(omega[i])
        # Apply the M step and obtain an update of M
        for j in range(K):
            m2 = np.sum(omega[:, j])
            m1 = np.sum(omega[:, j].reshape(-1, 1) * X, axis=0)
            # print(m1)
            M[j] =  m1 / m2
        # Calculate the NLL vale
        # print(omega.shape)
        tempn = []
        for i in range(n):
            temp = []
            for j in range(K):
                temp.append(ex_neg_norm(X[i], M[j], sigma))
            tempK = np.sum(temp)
            tempn.append(np.log(tempK))
            
        value[t] = -np.sum(tempn)

    return M, value

#Load the CC General dataset by pandas and keep the numerical part as a
# numpy array X
df = pd.read_csv('CC GENERAL.csv')
X = df.select_dtypes(include=np.number).fillna(0).values

```

```python
# plt.show()
# Normalize the data
X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
# plt.plot(X)
plt.show()
# Apply EM_clustering(X,sigma,K)
sigma = 9
K = 3
M, value = EM_clustering(X, sigma, K)

# Plot value vs iterations
plt.plot(value)
plt.xlabel('Iterations')
plt.ylabel('NLL Value')
plt.title('NLL Value vs Iterations')
plt.savefig('Code_2_2_iter.png')
plt.show()

```

### <center> Q3. (Bregman divergence)

#### 1.
For $\psi(x) = \lVert x \rVert^2_2$ the gradient of $\psi(x)$ is 
$$
\nabla \psi(x) = 2x
$$
Substituting this and $\psi(x) = \lVert x\rVert^2_2$, $\psi(y) = \lVert y\rVert^2_2$   into the Bregman divergence formula, we get:
$$
D_\psi(y, x) = \lVert y \rVert^2_2 - \lVert x \rVert^2_2 - 2x^T(y - x)
$$
Expanding the Euclidean norms and the gradient term:
$$
D_\psi(y, x) =  y^Ty - x^Tx - 2x^Ty + 2x^Tx
$$
Simplifying and combining terms
$$
D_\psi(y, x) =  y^Ty - 2x^Ty + x^Tx
$$
This then simplifies to the Euclidian distance between $y$ and $x$:
$$
D_\psi(y, x) = \lVert y - x \rVert^2_2
$$

#### 2.
The negative entropy function defined as: 
$$\psi(x) = \sum_{i} x_i \ln x_i$$

The first derivative of $\psi$ using product rule is:
$$\frac{\partial \psi}{\partial x_i} = \sum_i\ln{x_i}+1$$

The second derivative of $\psi$ is then:
$$\frac{\partial^2 \psi}{\partial x_i^2}  = \sum_i \frac{1}{x_i}$$

The mixed partial derivatives are zero in this case as the sum only contains $x_i$
$$\frac{\partial^2 \psi}{\partial x_i \partial x_j}  = \frac{\partial^2 \psi}{\partial x_j \partial x_i}  =  0$$

The Hessian matrix is defined as:
$$
H = \begin{bmatrix}
\frac{\partial^2 f}{\partial x_1^2} & \frac{\partial^2 f}{\partial x_1 \partial x_2} & \cdots & \frac{\partial^2 f}{\partial x_1 \partial x_n} \\
\frac{\partial^2 f}{\partial x_2 \partial x_1} & \frac{\partial^2 f}{\partial x_2^2} & \cdots & \frac{\partial^2 f}{\partial x_2 \partial x_n} \\
\vdots & \vdots & \ddots & \vdots \\
\frac{\partial^2 f}{\partial x_n \partial x_1} & \frac{\partial^2 f}{\partial x_n \partial x_2} & \cdots & \frac{\partial^2 f}{\partial x_n^2}
\end{bmatrix}
$$
Forming the Hessian therefore we have:
$$
H = \begin{bmatrix}
\frac{1}{x_1} & 0 & \cdots & 0 \\
0 & \frac{1}{x_2} & \cdots & 0\\
\vdots & \vdots & \ddots & \vdots \\
0 & 0 & \cdots & \frac{1}{x_n}
\end{bmatrix}
$$
The determinant of a diagonal matrix is the product of its diagonal elements. Therefore, the determinant of the Hessian matrix is 
$$
\det{H} =\prod_{i=1}^{n}\frac{1}{​x_i}
$$
which is always positive for $x_i \geq 0$
Hence, $\psi(x)$ is strictly convex.



#### 3.
For the negative entropy function, $D_\psi(y, x) = \sum_{i} x_i - \sum_{i} y_i + \sum_{i} y_i \ln \left( \frac{y_i}{x_i} \right)$
Substituting $\psi(x) = \sum_{i} x_i \ln x_i$ and $\nabla \psi(x) = 1 + \ln x$ into the Bregman divergence formula, we get:
$$
D_\psi(y, x) = \sum_{i} y_i \ln y_i - \sum_{i} x_i \ln x_i - \sum_{i} (1 + \ln x_i)(y_i - x_i)
$$
Expanding the last term:
$$
D_\psi(y, x) = \sum_{i} y_i \ln y_i - \sum_{i} x_i \ln x_i - \sum_{i} (y_i - x_i) - \sum_{i}y_i\ln x_i + \sum_{i} x_i\ln x_i
$$
Combining like terms:
$$
D_\psi(y, x) = \sum_{i} y_i (\ln y_i - \ln x_i) - \sum_{i} (y_i - x_i) 
$$
Simplifying this, we get:
$$
D_\psi(y, x) = \sum_{i} x_i - \sum_{i} y_i + \sum_{i} y_i \ln \left( \frac{y_i}{x_i} \right)
$$

If $\sum_{i} x_i = \sum_{i} y_i = 1$ (i.e., $x$ and $y$ are probability mass functions), then this simplifies to the Kullback-Leibler (KL) divergence. $D_{KL}(p(x)||q(x)) = \sum_{x} p(x) \ln \left( \frac{p(x)}{q(x)} \right)$
$$
D_\psi(y, x) = \sum_{i} y_i\ln \left( \frac{y_i}{x_i} \right)
$$
